# Calculate the reaction rate constant of a chemical reaction at a given temperature
2CPD -> DCPD

```python
from ThermoCR.QMthermo.qm_thermo import qm_thermo
from ThermoCR.QMkinetics.qm_kinetics import k_TST
from ThermoCR.tools import read_imaginary_vib

CPD_thermo = qm_thermo(atom_coord_path='CPD.out', ee=-194.283781614283, T=298.15, P=100000, verbose=False)
TS_thermo = qm_thermo(atom_coord_path='TS_CPD_DCPD.out', ee=-388.546183691360, T=298.15, P=100000, verbose=False)
DCPD_thermo = qm_thermo(atom_coord_path='DCPD.out', ee=-388.612084245544, T=298.15, P=100000, verbose=False)

delta_G = TS_thermo['G/(J/mol)'] - 2 * CPD_thermo['G/(J/mol)']
imaginary_freq = read_imaginary_vib('TS_CPD_DCPD.out')

k = k_TST(delta_G=delta_G, delta_n=1, tunnelling_effect='wigner', imaginary_freq=imaginary_freq, T=298.15, sigma=1)
print(f'k is {k} (molecule/m^3)^(-delta_n) * s^-1')
```

# Calculate the reaction rate constants at a series of temperatures

```python
from ThermoCR.QMthermo.qm_thermo import qm_thermo_scan
from ThermoCR.QMkinetics.qm_kinetics import k_TST_scan
from ThermoCR.tools import read_imaginary_vib

T = range(200, 750, 50)
imaginary_freq = read_imaginary_vib('TS_CPD_DCPD.out')
CPD_thermo_scan = qm_thermo_scan(atom_coord_path='CPD.out', ee=-194.283781614283, T=T, P=[100000], out_path='QMthermoScan_CPD.xlsx')
TS_thermo_scan = qm_thermo_scan(atom_coord_path='TS_CPD_DCPD.out', ee=-388.546183691360, T=T, P=[100000], out_path='QMthermoScan_TS.xlsx')
DCPD_thermo_scan = qm_thermo_scan(atom_coord_path='DCPD.out', ee=-388.612084245544, T=T, P=[100000], out_path='QMthermoScan_DCPD.xlsx')

k_scan_df = k_TST_scan(thermo_ts_path='QMthermoScan_TS.xlsx', 
                    thermo_r1_path='QMthermoScan_CPD.xlsx', thermo_r2_path='QMthermoScan_CPD.xlsx', 
                    thermo_p_path='QMthermoScan_DCPD.xlsx',
                    tunnelling_effect='eckart',
                    imaginary_freq=imaginary_freq, sigma=1, out_path='TST_scan_2CPD_to_DCPD.xlsx')

k_scan_df
```

# Fit the rate constants at different temperatures using the Arrhenius equation

```python
from ThermoCR.QMkinetics.fit_kinetics import fit_kinetics_model

fit_result = fit_kinetics_model(model_type='Arrhenius', data_path='TST_scan_2CPD_to_DCPD.xlsx', r_name_list=['CPD', 'CPD'],
                                p_name_list=['DCPD'], output_dir='fit_Arrhenius_result',)
```

# Calculate the rate constants at a series of temperatures based on VTST

```python
import pandas as pd
from ThermoCR.QMthermo.qm_thermo import qm_thermo_scan
from ThermoCR.QMkinetics.qm_kinetics import k_VTST_scan
from ThermoCR.tools import read_imaginary_vib

T = range(200, 750, 50)
imaginary_freq = read_imaginary_vib('TS_CPD_DCPD.out')
IRC_SP_df = pd.read_excel('IRC.xlsx')

IRC_thermo_scans = [
    qm_thermo_scan(atom_coord_path=f'./IRC_freq/{name}_freq.out', ee=sp, T=T, P=[100000], out_path=f'QMthermoScan_{name}.xlsx') 
    for name, sp in zip(IRC_SP_df['name'], IRC_SP_df['wB97Mp2'])
]

CPD_thermo_scan = qm_thermo_scan(atom_coord_path='CPD.out', ee=-194.283781614283, T=T, P=[100000], out_path='QMthermoScan_CPD.xlsx')
TS_thermo_scan = qm_thermo_scan(atom_coord_path='TS_CPD_DCPD.out', ee=-388.546183691360, T=T, P=[100000], out_path='QMthermoScan_TS.xlsx')
DCPD_thermo_scan = qm_thermo_scan(atom_coord_path='DCPD.out', ee=-388.612084245544, T=T, P=[100000], out_path='QMthermoScan_DCPD.xlsx')

k_vtst_scan = k_VTST_scan(
    thermo_irc_path_list=[f'QMthermoScan_{name}.xlsx' for name in IRC_SP_df['name']],
    thermo_r1_path='QMthermoScan_CPD.xlsx', thermo_r2_path='QMthermoScan_CPD.xlsx', 
    thermo_p_path='QMthermoScan_DCPD.xlsx',
    tunnelling_effect='eckart',
    imaginary_freq=imaginary_freq, sigma=1, out_path='VTST_scan_2CPD_to_DCPD.xlsx'
)

k_vtst_scan
```
